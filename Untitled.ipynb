```python
import pandas as pd
df=pd.read_csv("spotify.csv",index_col=0)#file_name
dff=[tuple(map(str,i)) for i in df.to_records()]
dff
import mysql.connector

mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="system",
  database="ak" # database name
)

mycursor = mydb.cursor()
mycursor


mycursor.execute("create table if not exists spotify (S_no varchar(50),Last_Updated varchar(50),Artist_Name varchar(50),Lead_Streams varchar(50),Feats varchar(50),Tracks varchar(50),One_Billion varchar(50),100Million varchar(50))") #table name and column name with data type
sqll = "INSERT INTO spotify VALUES (%s, %s, %s, %s, %s, %s, %s, %s)" #table anme with no of %s
val = dff
mycursor.executemany(sqll,val)

mydb.commit()
mydb.close()

print(mycursor.rowcount, "was inserted.")
```
